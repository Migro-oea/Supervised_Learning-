#**Iris Dataset**: *Classification of Flowers Based on Sepal Lenght*

```python
# Import necessary libraries for data manipulation and modeling
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeRegressor
```

Loading Iris Dataset and Creating Pandas Dataframe

```python
# Load the iris dataset
iris = load_iris()
# Create a pandas DataFrame from the dataset
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
# Display descriptive statistics of the DataFrame
display(iris_df.describe())
```

## Defining Features and Target Variable


Target Variable and the Features



```python
# Select the 'sepal length (cm)' column as the target variable for the regression model.
y = iris_df.loc[:, 'sepal length (cm)']
y.describe()
```

 Defining the Features and Create the Feature DataFrame `X`.



```python
# Define the feature columns to be used for the model.
feature_names = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)']
# Create a new DataFrame 'X' containing only the selected feature columns.
X = iris_df[feature_names]

X.describe()
```

## Model Training

## Model Training

Model Training Code


```python
# Define the Decision Tree Regressor model with a random state for reproducibility.
iris_df_model = DecisionTreeRegressor(random_state=1)
# Train the model using the training data (features X and target y).
iris_df_model.fit(X, y)

iris_df_model
```

**Making Predicions**

Prediction Code


```python
# Use the trained model to make predictions on the feature data X.
predictions = iris_df_model.predict(X)
# Print the generated predictions.
print(predictions)
```

## Comparing Preditions to Actuals

Comparison Code


```python
# Create a DataFrame to compare the actual target values (y) with the predicted values.
comparison = pd.DataFrame({'Actual': y, 'Predicted': predictions})
# Display the first few rows of the comparison DataFrame to show a sample of actual vs predicted values.
comparison.head()
```

## Model Validation


```python
# Import the function to split data into training and validation sets.
from sklearn.model_selection import train_test_split

# Split the feature and target data into training and validation sets.
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
# Create a new Decision Tree Regressor model for validation.
iris_df_model = DecisionTreeRegressor()
# Train the validation model on the training data.
iris_df_model.fit(train_X, train_y)
```

```python
# Make predictions on the validation features using the trained validation model.
val_predictions = iris_df_model.predict(val_X)
# Print the first 5 validation predictions.
print((val_predictions[:5]))
# Print the first 5 actual validation values for comparison.
print((val_y[:5]))
```
